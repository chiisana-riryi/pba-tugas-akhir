```python
import torch
from transformers import BertTokenizer
import pandas as pd
from transformers.utils import is_torch_available

print(is_torch_available())
```

# 2. Tokenise

```python
model = 'indobenchmark/indobert-base-p2'
tokenizer = BertTokenizer.from_pretrained(model)

aspek = "pelayanan"
review = "Pelayanannya sangat ramah dan cepat."

# Tokenisasi pasangan aspek-review
encoding = tokenizer(
    aspek,
    review,
    padding='max_length',
    truncation=True,
    max_length=20,
    return_tensors='pt'
)

print("input_ids:")
print(encoding['input_ids'])

print("\nattention_mask:")
print(encoding['attention_mask'])

print("\ntoken_type_ids:")
print(encoding['token_type_ids'])
```
